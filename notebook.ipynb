```python
from datasets import load_from_disk

# Load merged dataset
merged = load_from_disk("data_processed")
print("📊 MERGED DATASET STRUCTURE:")
print(f"Splits: {list(merged.keys())}")
print(f"Train: {len(merged['train']):,} samples")
print(f"Validation: {len(merged['validation']):,} samples")
print(f"Test: {len(merged['test']):,} samples")

# Check for empty responses
print("\n✅ EMPTY RESPONSE CHECK:")
for split in ["train", "validation", "test"]:
    empty_count = sum(1 for i in range(min(500, len(merged[split]))) if merged[split][i].get("response", "") == "")
    print(f"  {split}: {empty_count}/500 empty responses")

# Show task distribution
print("\n📋 TASK DISTRIBUTION (first 500 samples):")
from collections import Counter
task_counts = Counter()
for i in range(min(500, len(merged['train']))):
    task_counts[merged['train'][i]['task']] += 1
for task, count in sorted(task_counts.items()):
    print(f"  {task}: {count}")

# Sample from each task
print("\n🔍 SAMPLE PROMPTS & RESPONSES:")
seen_tasks = set()
for i in range(len(merged['train'])):
    task = merged['train'][i]['task']
    if task not in seen_tasks:
        sample = merged['train'][i]
        print(f"\n  Task: {task}")
        print(f"  Prompt (first 100 chars): {sample['prompt'][:100]}...")
        print(f"  Response: '{sample['response']}'")
        seen_tasks.add(task)
        if len(seen_tasks) == 6:
            break
```
